```python
from tensorboard.backend.event_processing import event_accumulator
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
root_dir = '/home/studio-lab-user/701-simplicity-bias/output/training_logs_221030_023412'
input_dims = []
train_acc = []
val_acc = []
srandom_acc = []
scrandom_acc = []
for dir in os.listdir(root_dir):
    if '7slabs_inputdim' not in dir or 'ref' in dir: continue
    input_dims.append(int(dir.split('_')[-1]))
    event_name = [fname for fname in os.listdir(os.path.join(root_dir, dir)) if fname.startswith('events.')][0]
    ea = event_accumulator.EventAccumulator(os.path.join(root_dir, dir, event_name),
          size_guidance={
              event_accumulator.COMPRESSED_HISTOGRAMS: 500,
              event_accumulator.IMAGES: 4,
              event_accumulator.AUDIO: 4,
              event_accumulator.SCALARS: 0,
              event_accumulator.HISTOGRAMS: 1,
          })
    ea.Reload()
    train_acc.append(ea.Scalars('Train/Accuracy')[-1].value)
    val_acc.append(ea.Scalars('Val/Accuracy')[-1].value)
    srandom_acc.append(ea.Scalars('s_randomized/Accuracy')[-1].value)
    scrandom_acc.append(ea.Scalars('sc_randomized/Accuracy')[-1].value)
lw, alpha = 1.5, 0.8
plt.plot(input_dims, train_acc, label='Training Accuracy', linewidth=lw, alpha=alpha)
plt.plot(input_dims, val_acc, label='Validation Accuracy', linewidth=lw, alpha=alpha)
# plt.plot(input_dims, srandom_acc, label='$S$-Randomized Accuracy', linewidth=lw, alpha=alpha)
# plt.plot(input_dims, scrandom_acc, label='$S^C$-Randomized Accuracy', linewidth=lw, alpha=alpha)
plt.xlabel('Number of input dimensions')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('/home/studio-lab-user/701-simplicity-bias/output/7slab_inputdim_ref.pdf', bbox_inches='tight', pad_inches=0, dpi=300)
plt.show()
```
